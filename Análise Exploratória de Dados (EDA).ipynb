Análise Exploratória de Dados (EDA) - Dataset Iris

Objetivo

Este trabalho tem como objetivo realizar uma Análise Exploratória de Dados (EDA) utilizando o dataset Iris.

A Análise Exploratória de Dados permite compreender as características do conjunto de dados através de medidas estatísticas e visualizações gráficas.

Durante esta análise serão estudados:

- Estrutura dos dados
- Tipos de variáveis
- Medidas de tendência central
- Medidas de dispersão
- Distribuição dos dados
- Correlação entre variáveis
- Comparação entre espécies de flores

O dataset Iris é amplamente utilizado em estudos de Estatística, Ciência de Dados e Machine Learning.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

# Carregamento do Dataset

Nesta etapa será realizado o carregamento do arquivo iris.csv para análise.

```python
df = pd.read_csv('iris.csv')

df.head()
```

# Visualização Inicial dos Dados

O método head() exibe as primeiras linhas do conjunto de dados.

Isso permite verificar:

- Estrutura da tabela
- Nomes das colunas

```python
df.shape
```

# Dimensões do Dataset

O método shape retorna:

- Quantidade de linhas (observações)
- Quantidade de colunas (variáveis)

Cada linha representa uma flor observada.

Cada coluna representa uma característica da flor.

```python
df.info()
```

# Informações Gerais

O método info() permite verificar:

- Quantidade de registros
- Tipo de cada variável
- Existência de valores ausentes

Essa análise é importante para identificar problemas nos dados antes das análises estatísticas.

```python
df.isnull().sum()
```

# Valores Ausentes

Valores ausentes (missing values) podem prejudicar análises estatísticas.

Nesta etapa verificamos se existem dados faltando em alguma coluna.

```python
df.describe()
```

# Estatística Descritiva

O método describe() apresenta:

- Média
- Desvio padrão
- Valor mínimo
- Quartis
- Valor máximo

Essas medidas fornecem uma visão geral sobre a distribuição dos dados.

# Medidas de Tendência Central

As medidas de tendência central indicam onde os dados tendem a se concentrar.

Serão analisadas:

- Média
- Mediana
- Moda

```python
df.mean(numeric_only=True)
```

## Média

A média representa o valor médio dos dados.

É obtida somando todos os valores e dividindo pela quantidade de observações.

```python
df.median(numeric_only=True)
```

## Mediana

A mediana representa o valor central dos dados ordenados.

Metade dos valores está abaixo dela e metade acima.

```python
df.mode()
```

## Moda

A moda é o valor que aparece com maior frequência no conjunto de dados.

# Medidas de Dispersão

As medidas de dispersão indicam o quanto os dados estão espalhados em torno da média.

Serão analisadas:

- Variância
- Desvio padrão

```python
df.var(numeric_only=True)
```

## Variância

A variância mede o grau de dispersão dos dados em relação à média.

Quanto maior a variância, maior a dispersão dos valores.

```python
df.std(numeric_only=True)
```

## Desvio Padrão

O desvio padrão é a medida de dispersão mais utilizada.

Ele indica o quanto os dados variam em torno da média.

Valores elevados indicam maior variabilidade.

# Histogramas

Os histogramas permitem visualizar a distribuição dos dados.

Eles mostram quais valores aparecem com maior frequência.

```python
df.hist(figsize=(12,8))
plt.show()
```

# Interpretação dos Histogramas

Através dos histogramas é possível observar:

- Concentração dos dados
- Assimetria
- Possíveis padrões de distribuição

# Boxplots

Os boxplots permitem visualizar:

- Mediana
- Quartis
- Valores mínimos
- Valores máximos
- Possíveis outliers

```python
for coluna in df.select_dtypes(include=np.number).columns:

    plt.figure(figsize=(8,4))

    sns.boxplot(x=df[coluna])

    plt.title(f'Boxplot - {coluna}')

    plt.show()
```

# Correlação

A correlação mede a relação entre duas variáveis.

Valores próximos de:

+1 → forte correlação positiva

0 → ausência de correlação

-1 → forte correlação negativa

```python
correlacao = df.select_dtypes(include=np.number).corr()

correlacao
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    correlacao,
    annot=True,
    cmap='coolwarm'
)

plt.show()
```

# Heatmap de Correlação

O heatmap facilita a identificação das relações entre variáveis.

Quanto mais próximo de 1, mais forte a relação positiva.

Quanto mais próximo de -1, mais forte a relação negativa.

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='petal.length',
    y='petal.width',
    hue='variety'
)

plt.title('Comprimento x Largura da Pétala')
plt.xlabel('Comprimento da Pétala')
plt.ylabel('Largura da Pétala')

plt.show()
```

Observa-se uma separação visual entre as espécies de flores.

A espécie Setosa apresenta um agrupamento claramente distinto das demais.

As espécies Versicolor e Virginica apresentam alguma sobreposição, porém ainda é possível identificar diferenças entre seus padrões.

```python
sns.pairplot(
    df,
    hue='variety'
)

plt.show()
```

O Pairplot permite visualizar simultaneamente a relação entre todas as variáveis do conjunto de dados.

Os gráficos indicam que as medidas das pétalas possuem maior capacidade de distinguir as espécies quando comparadas às medidas das sépalas.

A espécie Setosa apresenta uma separação bastante evidente das demais.

# Comparação Entre Espécies

Os gráficos de dispersão permitem observar se as espécies apresentam características distintas.

Essa análise é importante para verificar a separação natural entre grupos.

```python
df.groupby('variety').mean()
```

A análise das médias demonstra diferenças significativas entre as espécies.

A espécie Virginica apresenta, em média, os maiores valores de comprimento e largura de pétala.

A espécie Setosa apresenta os menores valores para essas características.

O boxplot evidencia diferenças entre as distribuições das espécies.

Observa-se que a espécie Virginica possui maiores comprimentos de pétala, enquanto a espécie Setosa apresenta os menores valores.

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x='variety',
    y='petal.length'
)

plt.title('Comprimento da Pétala por Espécie')

plt.show()
```

# Conclusão

O dataset Iris possui 150 observações referentes a diferentes espécies de flores distribuídas em três espécies: Setosa, Versicolor e Virginica.

A análise exploratória permitiu identificar:

- Estrutura dos dados
- Ausência ou presença de valores faltantes
- Medidas de tendência central
- Medidas de dispersão
- Distribuição das variáveis
- Correlação entre características

Os gráficos demonstraram que algumas variáveis possuem forte relação entre si, principalmente aquelas relacionadas às pétalas.

Também foi possível observar diferenças entre as espécies analisadas, evidenciando padrões que podem ser utilizados em tarefas de classificação e aprendizado de máquina.

Não foram identificados valores ausentes no conjunto de dados.

As análises demonstraram forte correlação positiva entre o comprimento e a largura das pétalas.

Os histogramas permitiram observar a distribuição das variáveis, enquanto os boxplots auxiliaram na identificação da dispersão e de possíveis valores extremos.

Os gráficos de dispersão e o pairplot evidenciaram que as características relacionadas às pétalas possuem maior capacidade de distinguir as espécies.

Portanto, conclui-se que as medidas das pétalas são as variáveis mais relevantes para a diferenciação das espécies presentes no dataset Iris.

# Frequência das Espécies

A frequência representa a quantidade de ocorrências de cada categoria presente na variável.

Nesta etapa será analisada a distribuição das espécies de flores presentes no dataset.

```python
df['variety'].value_counts()
```

# Interpretação da Frequência

A tabela apresenta a quantidade de flores pertencentes a cada espécie.

Essa análise permite verificar se existe equilíbrio entre as classes do conjunto de dados.

```python
df['variety'].value_counts(normalize=True) * 100
```

# Frequência Relativa (%)

A frequência relativa mostra a porcentagem de cada espécie em relação ao total de observações.

Essa medida facilita a comparação proporcional entre os grupos.

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='variety'
)

plt.title('Quantidade de Flores por Espécie')
plt.xlabel('Espécie')
plt.ylabel('Quantidade')

plt.show()
```

# Interpretação do Gráfico de Barras

O gráfico demonstra a distribuição das espécies presentes no dataset.

Observa-se que as três espécies possuem a mesma quantidade de registros.

Isso indica que o conjunto de dados é balanceado, ou seja, nenhuma espécie possui vantagem numérica sobre as demais.

```python
frequencia = df['variety'].value_counts()

plt.figure(figsize=(8,8))

plt.pie(
    frequencia,
    labels=frequencia.index,
    autopct='%1.1f%%'
)

plt.title('Distribuição das Espécies')

plt.show()
```

# Interpretação do Gráfico de Pizza

O gráfico de pizza confirma a distribuição equilibrada das espécies.

Cada espécie representa aproximadamente um terço do conjunto de dados.

Esse equilíbrio é importante para análises estatísticas e modelos de aprendizado de máquina.

```python
df.groupby('variety').agg(
    ['mean','median','std']
).round(2)
```

A tabela apresenta:

- Média

- Desvio padrão

para cada espécie.

Observa-se que as espécies apresentam características distintas principalmente nas variáveis relacionadas às pétalas.

A espécie Virginica apresenta os maiores valores médios de comprimento e largura de pétala.

A espécie Setosa apresenta os menores valores médios.

Essas diferenças ajudam a explicar a separação observada nos gráficos de dispersão e no pairplot.

A principal descoberta da Análise Exploratória de Dados (EDA):

A espécie Setosa apresentou separação muito clara das demais, enquanto Versicolor e Virginica apresentaram algumas características semelhantes. Além disso, observou-se forte correlação positiva entre comprimento e largura das pétalas, e o conjunto de dados mostrou-se balanceado, sem valores ausentes.
